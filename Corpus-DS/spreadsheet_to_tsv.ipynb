```python
import pandas as pd
```

```python
df = pd.read_csv('DS.tsv', sep='\t')
df.head()
```

```python
df0 = df[['Python', '日本語']]
df0.head()
```

```python
df1 = df[['Python', '日本語.1']]
df1 = df1.rename(columns={'日本語.1': '日本語'})
df1.head()
```

```python
df2 = df[['Python', '日本語.2']]
df2 = df2.rename(columns={'日本語.2': '日本語'})
df2.head()
```

```python
df3 = df[['Python', '日本語.3']]
df3 = df3.rename(columns={'日本語.3': '日本語'})
df3.head()
```

```python
df4 = df[['Python', '日本語.4']]
df4 = df4.rename(columns={'日本語.4': '日本語'})
df4.head()
```

```python
df_new = pd.concat([df0, df1, df2, df3, df4])
df_new.dropna()
```

```python
df_new = df_new.dropna()
df_new = df_new.reset_index(drop=True)
df_new
```

```python
df_new.to_csv('ds.tsv', index=False, sep='\t')
```
